```python
scaler = StandardScaler().fit_transform(df)
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
```

```python
pca = PCA(n_components = 2)
df_1 = pca.fit_transform(scaler)
df_1 = pd.DataFrame(df_1, columns = ['P1', 'P2'])
df_1
```

```python
dbscan = DBSCAN(eps = 0.3, min_samples = 6).fit(df_1)
label = dbscan.labels_
label
```

```python
plt.figure(figsize = (7, 8))
sns.scatterplot(x = df_1['P1'], y = df_1['P2'], data = 'df_1', hue = label)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = []
sil = []
for i in range(2, 16):
    km = KMeans(n_clusters = i, init = 'k-means++', random_state = 42).fit(df_1)
    wcss.append(km.inertia_)
    sil.append(silhouette_score(df_1, km.labels_))
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.plot(range(2, 16), wcss)
plt.show()

plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 2)
plt.plot(range(2, 16), sil)
plt.show()
```

```python
k_means = KMeans(n_clusters = 7).fit(df_1)
k_means_pred = k_means.predict(df_1)
```

```python
centers = k_means.cluster_centers_
centers
```

```python
df_1['cluster'] = k_means_pred
plt.figure(figsize=(20, 10))
sns.scatterplot(data = df_1, x = 'P1', y = 'P2', hue = 'cluster', s = 100)
plt.show()
```
